# 3. Correlation matrix heatmap

```python
# Create a subset excluding variables not needed in analysis

df_sub = df[['Happiness Rank', 'Happiness Score', 'Economy GDP per Capita', 'Family Social Support','Health Life Expectancy', 'Freedom to Make Choices', 'Perceptions of Corruption', 'Generosity', 'COVID-19 deaths per 100,000 population in 2020']]
df_sub
```

```python
# Create Correlation Matrix

corr_matrix = df_sub.corr()
corr_matrix
```

```python
# Set Up Mask To Hide Upper Triangle

mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)]= True
mask
```

```python
# Create Heatmap with Seaborn

f, ax = plt.subplots(figsize=(11, 15)) 
heatmap = sns.heatmap(corr_matrix, 
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap = 'coolwarm',
                      cbar_kws = {'shrink': .4, 
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1, 
                      vmax = 1,
                      annot = True,
                      annot_kws = {'size': 12})
#add the column names as labels
ax.set_yticklabels(corr_matrix.columns, rotation = 0)
ax.set_xticklabels(corr_matrix.columns)
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
```

#### Looking at the heat map we can see that Happiness Score & Rank have a direct relationship, which makes sense. We can also see which of the 6 pillars are most correlated with Score/Rank. This is the most interesting facet as Corruption and Generosity are greatly different than the other pillars.

```python
# Export Heatmap

heatmap.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'corr_heatmap.png'), bbox_inches = 'tight')
```

# 6. Categorical plot

```python
# Use a histogram to visualize the distribution of the variables. 

sns.histplot(df['Happiness Score'], bins = 10, kde = True)
```

```python
# Create happiness categories

df.loc[df['Happiness Score'] < 4, 'Happiness Category'] = 'Low'
df.loc[df['Happiness Score'] > 6.5, 'Happiness Category'] = 'High'
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] <= 6.5), 'Happiness Category'] = 'Middle'
```

```python
df['Happiness Category'].value_counts(dropna=False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the happiness categories created above

sns.set(style='ticks')
g = sns.catplot(x='Happiness Score', y='Perceptions of Corruption', hue='Happiness Category', data=df)
```

```python
sns.set(style='ticks')
g = sns.catplot(x='Happiness Score', y='Generosity', hue='Happiness Category', data=df)
```

### These categorical plots have helped greatly over the scatter plots. We can see that these two pillars are biased. Happy people tend to hold these pillars at a higher regard than there counterpoints.

```python
#Looking at Deaths

sns.set(style='ticks')
g = sns.catplot(x='Happiness Score', y='COVID-19 deaths per 100,000 population in 2020', hue='Happiness Category', data=df)
```
